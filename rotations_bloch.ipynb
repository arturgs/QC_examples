# Bloch rotation fundamentals

## import Qutip package

```python
from qutip import *
from qutip.qip import *
```

## define Block sphere object

```python
 b = Bloch()
```

## import Pauli matrices

```python
sx = sigmax()
sy = sigmay()
sz = sigmaz()
```

### We create a Quantum state, and operate with a Pauli operator. We plot the result on the Bloch sphere

```python
vec = basis(2,0)
a = sx * vec
b.add_states(a)
b.show()
```

### we clean the vectors already in the Bloch sphere

```python
b.clear()
```

## Rotations
### we define a rotation along some axis, which is applied to a given Quantum state

```python
Ry = ry(0.5)
vec = basis(2,0)
a = Ry * vec
Rx = rx(0.5)
a = Rx * a
b.add_states(a)
b.show()
```
